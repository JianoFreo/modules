# ðŸ§  Python OOP: Classes, Objects, and `self` â€” Big Picture First

Python lets you **group data and functions together**.  
That group is called a **class**.  

From a class, Python can create real things you can use.  
Those real things are called **objects** (also called **instances**).

ðŸ‘‰ **`self` is how an object talks about itself.**

That's it. Everything else is detail.

## 1ï¸âƒ£ What is a Class?

Think of a class as a **blueprint**.

**Real life example:**  
- A house blueprint  
- It describes what a house *has* and what it *can do*

> â— You cannot pet a blueprint. You need a real dog.

````python
class Dog:
    pass

# This does nothing yet.
# It only describes what a Dog IS.
````

## 2ï¸âƒ£ What is an Object / Instance?

An **object** (or **instance**) is a **real thing** made from a class.

- `Dog` = blueprint  
- `my_dog` = real dog  

You can make **many objects** from one class â€” each is a separate real thing.

````python
my_dog = Dog()

dog1 = Dog()
dog2 = Dog()

print(dog1)
print(dog2)
print(dog1 is dog2)  # They are separate objects
````

## 3ï¸âƒ£ What is a Method?

A **method** is just a **function inside a class**.

- `bark` is a method  
- It belongs to the class  
- Objects can use it

````python
class Dog:
    def bark(self):
        print("Woof!")

dog1 = Dog()
dog1.bark()
````

## 4ï¸âƒ£ Soâ€¦ WHAT IS `self`? ðŸ”¥

`self` means: **"this specific object that is calling the method"**

When you write:
```python
dog1.bark()
```

Python secretly turns it into:
```python
Dog.bark(dog1)
```

ðŸ‘‰ `dog1` is passed in as `self`  
So inside the method: **`self` IS `dog1`**

## 5ï¸âƒ£ Why Python Needs `self`

````python
class Dog:
    def bark(self):
        print("Woof from", self)

dog1 = Dog()
dog2 = Dog()

dog1.bark()
dog2.bark()

# Same method, different objects.
# self lets Python know: "Which dog is barking?"
````

## 6ï¸âƒ£ What is `self.variable`?

`self.variable` means: **"Store something inside this object"**

Each object remembers its **own** data.

````python
class Dog:
    def set_name(self, name):
        self.name = name

dog1 = Dog()
dog2 = Dog()

dog1.set_name("Buddy")
dog2.set_name("Max")

print(dog1.name)  # Buddy
print(dog2.name)  # Max

# Same class. Different stored values.
````

## 7ï¸âƒ£ Why NOT just use normal variables?

| Code | Belongs to |
|------|------------|
| `name = "Buddy"` | âŒ The program (global/local scope) |
| `self.name = "Buddy"` | âœ… That specific object |

Objects carry their **own** data.

````python
# Without self.name â€” this breaks things
class BadDog:
    def set_name(self, name):
        name = name  # This is a local variable â€” lost after the method ends!

bad = BadDog()
bad.set_name("Ghost")
# print(bad.name)  # Uncomment this â€” it will raise AttributeError!
````

````python
# With self.name â€” works correctly
class GoodDog:
    def set_name(self, name):
        self.name = name  # Stored inside the object

good = GoodDog()
good.set_name("Rex")
print(good.name)  # Rex
````

## 8ï¸âƒ£ Your LeetCode Example Explained

```python
class Solution:
    def longestCommonPrefix(self, strs):
        ...
```

**What's happening:**

```python
Solution().longestCommonPrefix(strs)
```

Internally becomes:

```python
Solution.longestCommonPrefix(the_object, strs)
```

- `self` â†’ the object  
- `strs` â†’ input data  

You don't store `strs` in `self` because:  
- You only need it once  
- You don't need to remember it later

````python
class Solution:
    def longestCommonPrefix(self, strs):
        if not strs:
            return ""
        prefix = strs[0]
        for word in strs[1:]:
            while not word.startswith(prefix):
                prefix = prefix[:-1]
                if not prefix:
                    return ""
        return prefix

sol = Solution()
print(sol.longestCommonPrefix(["flower", "flow", "flight"]))  # fl
print(sol.longestCommonPrefix(["dog", "racecar", "car"]))      # ""
````

## 9ï¸âƒ£ When `self` is Required but Not Used

```python
def longestCommonPrefix(self, strs):
```

You **must** include `self` because Python always passes the object â€” even if you don't use it.

It's like a phone call where caller info is always included.  
You may ignore it, but it's always there.

````python
class Demo:
    def greet(self):          # self is required â€” even though we don't use it
        print("Hello!")

d = Demo()
d.greet()  # Works fine

# What happens if you forget self?
class Broken:
    def greet():              # Missing self!
        print("Hello!")

b = Broken()
# b.greet()  # Uncomment â€” raises TypeError: greet() takes 0 positional arguments but 1 was given
````

## ðŸ”Ÿ Simple Rules to Memorize

| Term | Meaning |
|------|---------|
| `self` | The current object |
| `self.x` | Data stored **inside** that object |
| Method parameters | **Temporary** data (used and gone) |
