### What is Binary Search?
**Binary search** finds target in sorted array by repeatedly dividing search space in half.

```
Array: [1, 3, 5, 7, 9, 11]
Find 7:  Left=0, Right=5
  Mid=2 (value=5) < 7 â†’ Left=3
  Mid=4 (value=9) > 7 â†’ Right=3
  Mid=3 (value=7) = 7 âœ“
```

**Key Properties:**
- O(log n) time complexity
- REQUIRES sorted input!
- Elegant with recursion or iteration
- Many variations for boundaries

````python
# Binary search - iterative
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    
    return -1  # Not found

arr = [1, 3, 5, 7, 9, 11]
print(f"Binary search for 7: {binary_search(arr, 7)}")  # 3
print(f"Binary search for 4: {binary_search(arr, 4)}")  # -1
````

### Time Complexity Analysis

| Algorithm | Time | Space | Note |
|-----------|------|-------|------|
| **Binary Search** | O(log n) | O(1) | Limited to sorted |
| **Linear Search** | O(n) | O(1) | Works on unsorted |

**Why O(log n)?**
- Each step eliminates half remaining elements
- 1000 elements â†’ 10 steps max

### Key Concepts & Patterns

#### 1. **Left Boundary Search**
Find leftmost position of target.

````python
# Find leftmost occurrence
def left_boundary(arr, target):
    left, right = 0, len(arr) - 1
    result = -1
    
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            result = mid
            right = mid - 1  # Keep searching left
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    
    return result

arr = [5, 7, 7, 7, 9, 10]
print(left_boundary(arr, 7))  # 1
````

### ðŸ”‘ Key Points Before Assessment

âœ… **Remember:**
1. Input MUST be sorted
2. O(log n) is very fast
3. Three variations: exact, left boundary, right boundary
4. Watch for integer overflow (use mid = left + (right-left)//2)
5. Check left <= right termination condition

âŒ **Avoid:**
- Using on unsorted arrays
- Off-by-one errors in boundaries
- Integer overflow with mid calculation
