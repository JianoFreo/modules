### What is a Queue?
A **Queue** is a FIFO (First-In-First-Out) data structure. First element added is first to be removed.

**Analogy:** Line at store - first person in line is first to checkout.

```
Enqueue 1: [1]
Enqueue 2: [1, 2]
Enqueue 3: [1, 2, 3]
Dequeue:   [2, 3]  (returns 1)
Front:     2
```

````python
from collections import deque

queue = deque()
queue.append(1)           # Enqueue
queue.append(2)
queue.append(3)
print(f"Queue: {queue}")
print(f"Dequeue: {queue.popleft()}")  # Remove 1
print(f"Front: {queue[0]}")
print(f"Size: {len(queue)}")
````

### Time Complexity

| Operation | Time | Using |
|-----------|------|-------|
| Enqueue | O(1) | append() |
| Dequeue | O(1) | popleft() |
| Peek | O(1) | queue[0] |

**Note:** Use deque, not list! (list.pop(0) is O(n))

### Key Queue Patterns

#### 1. **BFS (Breadth-First Search)**

````python
def bfs(start_node):
    queue = deque([start_node])
    visited = set([start_node])
    
    while queue:
        node = queue.popleft()
        print(node, end=" ")
        
        for neighbor in node.neighbors:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
````
