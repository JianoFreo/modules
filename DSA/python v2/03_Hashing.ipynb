### What is Hashing?
**Hashing** maps values to indices using a hash function. Provides O(1) average time lookup.

**Hash Function Example:**
```
hash("apple") â†’ 5
hash("banana") â†’ 7
hash(2) â†’ 2
```

**Key Properties:**
- Fast lookup: O(1) average case
- Collision handling via chaining or open addressing
- Good for frequency counting and membership tests
- Deterministic - same input always gives same hash

````python
# Hash map (dictionary) basics
d = {}
d['apple'] = 5
d['banana'] = 3

print(f"Dictionary: {d}")
print(f"d['apple']: {d['apple']}")
print(f"Keys: {d.keys()}")
print(f"Values: {d.values()}")
print(f"'apple' in d: {'apple' in d}")
print(f"d.get('grape', 0): {d.get('grape', 0)}")
````

### Time Complexity Analysis

| Operation | Average | Worst |
|-----------|---------|-------|
| **Lookup** | O(1) | O(n) |
| **Insert** | O(1) | O(n) |
| **Delete** | O(1) | O(n) |

**Worst Case:** When many collisions occur (poor hash function or many identical hashes).

**Load Factor:** ratio of items to slots. Keep < 0.75 for good performance.

### Key Concepts & Patterns

#### 1. **Frequency Counting**
Count occurrences of elements for quick lookups.

````python
from collections import Counter

# Frequency counting
arr = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
freq = Counter(arr)
print(f"Array: {arr}")
print(f"Frequency: {freq}")
print(f"Most common 2: {freq.most_common(2)}")

# Find majority element
majority = freq.most_common(1)[0][0]
print(f"Majority element: {majority}")
````

#### 2. **Two Sum with Hash Map**
Find two numbers summing to target in O(n) time.

````python
# Two sum using hash map
def two_sum(nums, target):
    seen = {}
    for i, num in enumerate(nums):
        complement = target - num
        if complement in seen:
            return [seen[complement], i]
        seen[num] = i
    return []

result = two_sum([2, 7, 11, 15], 9)
print(f"Two sum indices: {result}")  # [0, 1]

# Why it works:
# num=2: need 9-2=7, not seen yet, add 2â†’0
# num=7: need 9-7=2, found at index 0, return [0, 1]
````

#### 3. **Character Mapping**
Create bijective mapping between strings.

````python
# Isomorphic strings
def is_isomorphic(s, t):
    if len(s) != len(t):
        return False
    
    s_map = {}  # s â†’ t mapping
    t_map = {}  # t â†’ s mapping
    
    for c1, c2 in zip(s, t):
        if c1 in s_map:
            if s_map[c1] != c2:
                return False
        else:
            s_map[c1] = c2
        
        if c2 in t_map:
            if t_map[c2] != c1:
                return False
        else:
            t_map[c2] = c1
    
    return True

print(is_isomorphic("egg", "add"))      # True
print(is_isomorphic("foo", "bar"))     # False
````

#### 4. **Happy Number Detection**
Use set to detect cycle in repeated operations.

````python
# Happy number: repeatedly sum squares until 1 or cycle
def is_happy(n):
    seen = set()
    while n != 1 and n not in seen:
        seen.add(n)
        n = sum(int(digit) ** 2 for digit in str(n))
    return n == 1

print(f"19 is happy: {is_happy(19)}")  # True
print(f"2 is happy: {is_happy(2)}")    # False

# Trace for 19:
# 19 â†’ 1Â² + 9Â² = 1 + 81 = 82
# 82 â†’ 8Â² + 2Â² = 64 + 4 = 68
# 68 â†’ 6Â² + 8Â² = 36 + 64 = 100
# 100 â†’ 1Â² + 0Â² + 0Â² = 1 (HAPPY)
````

### Set vs Dictionary

| Operation | Set | Dict | Time |
|-----------|-----|------|------|
| Membership test | {1,2,3} | {1: 'a'} | O(1) avg |
| Count occurrences | Not good | Good | O(1) per lookup |
| Remove duplicates | Excellent | Need keys() | O(n) |
| Store relationships | No | Yes | O(1) lookup |

**Use Set for:** Membership, uniqueness
**Use Dict for:** Counting, mapping

````python
# Different tools for different jobs
arr = [1, 2, 2, 3, 3, 3]

# Remove duplicates â†’ use set
unique = set(arr)
print(f"Unique: {unique}")  # {1, 2, 3}

# Count occurrences â†’ use dict
freq = {}
for num in arr:
    freq[num] = freq.get(num, 0) + 1
print(f"Frequencies: {freq}")  # {1: 1, 2: 2, 3: 3}
````

### ðŸ”‘ Key Points Before Assessment

âœ… **Remember:**
1. Hash maps have O(1) average lookup
2. Use Counter for frequency problems
3. Use set for membership and uniqueness
4. Mappings can go both ways (bijection)
5. Collision handling impacts worst case
6. Python dict is ordered (Python 3.7+)

âŒ **Avoid:**
- Assuming O(1) in worst case (can be O(n) with many collisions)
- Forgetting to initialize default values
- Using unhashable types as keys (lists, dicts)

### What is Hashing?
**Hashing** is a technique to map data to a fixed-size value using a **hash function**. In Python, we use **dictionaries** and **sets** for hash-based operations.

**Hash Concept:**
```
Hash Function: key â†’ hash_value â†’ bucket
Example: 'name' â†’ hash(123) â†’ dictionary[123]
```

**Properties:**
- O(1) average access/insert/delete
- O(n) worst case (collision)
- Keys must be hashable (immutable)
- Unordered (Python 3.7+ maintains insertion order)
code

# Dictionary (Hash Map) Basics
hash_map = {}
hash_map['apple'] = 5
hash_map['banana'] = 3
hash_map['cherry'] = 8

print(f"'orange' in hash_map: {'orange' in hash_map}")

# Iterate
for key, value in hash_map.items():
    print(f"{key}: {value}")

| Delete | O(1) | O(n) |
| Search | O(1) | O(n) |

**Worst case occurs when:**
- Hash collision: multiple keys map to same value
- All keys hash to same bucket
