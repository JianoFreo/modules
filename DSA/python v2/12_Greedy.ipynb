### What is Greedy?
**Greedy** algorithm makes locally optimal choices, hoping for global optimum.

```
Coin change greedy (doesn't always work!):
coins = [1, 3, 4], amount = 6
Greedy: 4 + 1 + 1 = 3 coins
Optimal: 3 + 3 = 2 coins âœ—
```

**When it works:**
- Problem has optimal substructure
- Greedy choice leads to optimal solution
- Standard coins: greedy works!
