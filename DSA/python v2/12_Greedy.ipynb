### What is Greedy?
**Greedy algorithms** make locally optimal choices at each step, hoping to find a global optimum.

**When to use:**
- Problem has greedy choice property
- Problem has optimal substructure
- NOT always optimal (need to prove)

**Examples:**
- Huffman coding
- Activity selection
- Coin change
- Fractional knapsack

```python
# Example 1: Coin Change (minimize coins)
def min_coins(amount, coins):
    coins.sort(reverse=True)
    count = 0
    
    for coin in coins:
        while amount >= coin:
            amount -= coin
            count += 1
    
    return count if amount == 0 else -1

print(min_coins(17, [1, 5, 10]))  # 3 (10 + 5 + 2)
```

### Example 2: Activity Selection (Max activities)

```python
def activity_selection(start, end):
    # Sort by end time
    activities = sorted(zip(start, end), key=lambda x: x[1])
    
    selected = [activities[0]]
    last_end = activities[0][1]
    
    for i in range(1, len(activities)):
        if activities[i][0] >= last_end:
            selected.append(activities[i])
            last_end = activities[i][1]
    
    return selected
```
