### What is a Stack?
A **stack** (LIFO - Last In First Out) data structure where last inserted element is removed first.

```
Stack Operations:        Push(1)  Push(2)  Pop()  Top
                           |        |      â†’â†’    |
                         [1]      [1,2]    [1]   1
```

**Key Properties:**
- LIFO ordering - last element out first
- O(1) push and pop operations
- Used for recursion, undo/redo, expression evaluation
- Can use Python list with append/pop methods

````python
# Stack using list
stack = []
stack.append(1)        # Push
stack.append(2)
stack.append(3)
print(f"Stack: {stack}")

while stack:
    print(f"Pop: {stack.pop()}")  # Removes from end

# Output: Pop: 3, Pop: 2, Pop: 1
````

### Time Complexity Analysis

| Operation | Time |
|-----------|------|
| **Push** | O(1) |
| **Pop** | O(1) |
| **Peek (top)** | O(1) |
| **Search** | O(n) |

Python `list.append()` and `list.pop()` are O(1) amortized.

### Key Concepts & Patterns

#### 1. **Parentheses Matching**
Use stack to validate matching brackets.

````python
# Valid parentheses
def is_valid(s):
    stack = []
    pairs = {'(': ')', '[': ']', '{': '}'}
    
    for char in s:
        if char in pairs:     # Opening bracket
            stack.append(char)
        else:                 # Closing bracket
            if not stack or pairs[stack.pop()] != char:
                return False
    
    return len(stack) == 0  # All matched

print(is_valid("()"))        # True
print(is_valid("([{}])"))   # True
print(is_valid("([)]"))     # False (wrong order)
````

#### 2. **Monotonic Stack**
Maintain stack with elements in increasing/decreasing order.

````python
# Daily temperatures - find next warmer day
def daily_temperatures(temps):
    result = [0] * len(temps)
    stack = []  # Store indices
    
    for i, temp in enumerate(temps):
        while stack and temps[stack[-1]] < temp:
            prev_idx = stack.pop()
            result[prev_idx] = i - prev_idx  # Days to wait
        stack.append(i)
    
    return result

temps = [73, 74, 75, 71, 69, 72, 76, 73]
print(daily_temperatures(temps))
# Expected: [1, 1, 4, 2, 1, 1, 0, 0]
````
