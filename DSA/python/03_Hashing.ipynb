<span style="color:green">**H1: Two Sum**</span>
Find indices of two numbers that add to target. nums=[2,7,11,15], target=9.

**Output:** `[0, 1]`

```python
# H1: Find two numbers that add to target
nums = [2, 7, 11, 15]
target = 9

seen = {}
result = []

for i in range(len(nums)):
    complement = target - nums[i]
    
    if complement in seen:
        result = [seen[complement], i]
        break
    
    seen[nums[i]] = i

print(result)  # Output: [0, 1]
```
