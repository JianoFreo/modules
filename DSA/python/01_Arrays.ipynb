<span style="color:green">**A6: Prefix Sum Array**</span>
Create prefix sum array for [1, 2, 3, 4, 5].

**Output:** `[1, 3, 6, 10, 15]`

```python
array = [1,2,3,4,5]
total = []
running_sum = 0
for element in array:
    running_sum += element
    total.append(running_sum)
print(total)  # Output: [1, 3, 6, 10, 15]
```

<span style="color:green">**A7: Sum of Range**</span>
Find sum from index 1 to 3 in [1, 2, 3, 4, 5] using prefix sum.

**Output:** `9` (2+3+4)

```python
array = [1, 2, 3, 4, 5]
total = 0
for element in array[1:4]:  # indices 1 to 3 (2, 3, 4)
    total += element
print(total)  # Output: 9
```

<span style="color:green">**A8: Two Sum**</span>
Find two numbers that sum to target. arr=[2, 7, 11, 15], target=9.

**Output:** `[2, 7]`

```python
array = [2, 7, 11, 15]
two_nums = []
for num1 in array:
    for num2 in array:
        if num1 + num2 == 9 and num1 < num2:
            two_nums.append(num1)
            two_nums.append(num2)
    if len(two_nums) == 2:
        break
print(two_nums) 
```

<span style="color:green">**A9: Three Sum**</span>
Find three numbers that sum to 0 in [-1, 0, 1, 2, -1, -4].

**Output:** `[[-1, -1, 2], [-1, 0, 1]]`

```python
# Approach 1: Brute Force (O(n³))
list = [-1, 0, 1, 2, -1, -4]
nums = []
for i in range(len(list)):
    for j in range(i+1, len(list)):
        for k in range(j+1, len(list)):
            if list[i] + list[j] + list[k] == 0:
                triplet = sorted([list[i], list[j], list[k]])
                if triplet not in nums:  # Avoid duplicates
                    nums.append(triplet)
print(nums)  # Output: [[-1, -1, 2], [-1, 0, 1]]

# ===================================
```
