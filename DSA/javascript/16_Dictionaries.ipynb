## EASY (DC1-DC5)

<span style="color:green">**DC1: Count Frequency**</span>

Count elements.

**Input:** `["a", "b", "a", "c", "b", "a"]`

**Expected Output:** `{"a": 3, "b": 2, "c": 1}`

<span style="color:green">**DC2: Group by Value**</span>

Group items.

**Input:** `[{"name": "Alice", "role": "dev"}, {"name": "Bob", "role": "dev"}, {"name": "Carol", "role": "qa"}]`

**Expected Output:** `{"dev": ["Alice", "Bob"], "qa": ["Carol"]}`

<span style="color:green">**DC3: Key Exists**</span>

Check key.

**Input:** `{"name": "Alice", "age": 30}, key = "name"`

**Expected Output:** `true`

<span style="color:green">**DC4: Get Default**</span>

Safe get.

**Input:** `{"name": "Alice"}, key = "age", default = 0`

**Expected Output:** `0`

<span style="color:green">**DC5: Merge Dictionaries**</span>

Merge.

**Input:** `{"a": 1, "b": 2}, {"b": 3, "c": 4}`

**Expected Output:** `{"a": 1, "b": 3, "c": 4}`

## INTERMEDIATE (DC6-DC12)

<span style="color:green">**DC6: Invert Dictionary**</span>

Swap keys/values.

**Input:** `{"a": 1, "b": 2, "c": 3}`

**Expected Output:** `{1: "a", 2: "b", 3: "c"}`

<span style="color:green">**DC7: Sort by Value**</span>

Sort items.

**Input:** `{"apple": 3, "banana": 1, "cherry": 2}`

**Expected Output:** `[("banana", 1), ("cherry", 2), ("apple", 3)]`

<span style="color:green">**DC8: Common Keys**</span>

Find common.

**Input:** `{"a": 1, "b": 2, "c": 3}, {"a": 4, "c": 5, "d": 6}`

**Expected Output:** `{"a": 1, "c": 3}`

<span style="color:green">**DC9: Accumulate Values**</span>

Sum by key.

**Input:** `[{"item": "apple", "price": 1}, {"item": "apple", "price": 2}, {"item": "banana", "price": 3}]`

**Expected Output:** `{"apple": 3, "banana": 3}`

<span style="color:green">**DC10: Nested Access**</span>

Access nested.

**Input:** `{"user": {"name": "Alice", "age": 30}}, path = "user.age"`

**Expected Output:** `30`

<span style="color:green">**DC11: Filter Dictionary**</span>

Filter.

**Input:** `{"a": 1, "b": 2, "c": 3, "d": 4}, predicate = (val > 2)`

**Expected Output:** `{"c": 3, "d": 4}`

<span style="color:green">**DC12: Transform Values**</span>

Transform.

**Input:** `{"a": 1, "b": 2, "c": 3}, mapper = (val * 2)`

**Expected Output:** `{"a": 2, "b": 4, "c": 6}`

## ADVANCED (DC13-DC17)

<span style="color:green">**DC13: Flatten Dictionary**</span>

Flatten nested.

**Input:** `{"a": {"b": 1, "c": 2}, "d": 3}`

**Expected Output:** `{"a.b": 1, "a.c": 2, "d": 3}`

<span style="color:green">**DC14: Unflatten Dictionary**</span>

Unflatten.

**Input:** `{"a.b": 1, "a.c": 2, "d": 3}`

**Expected Output:** `{"a": {"b": 1, "c": 2}, "d": 3}`

<span style="color:green">**DC15: Difference Dictionaries**</span>

Find differences.

**Input:** `{"a": 1, "b": 2, "c": 3}, {"a": 1, "b": 5, "d": 4}`

**Expected Output:** `{"b": (2, 5), "c": (3, None), "d": (None, 4)}`

<span style="color:green">**DC16: Deep Copy**</span>

Deep copy nested.

**Input:** `{"a": {"b": [1, 2, 3]}, "c": 4}`

**Expected Output:** `{"a": {"b": [1, 2, 3]}, "c": 4}` (independent copy)`

<span style="color:green">**DC17: Path Finding**</span>

Find path to value.

**Input:** `{"a": {"b": {"c": 5}}, "d": 5}, target = 5`

**Expected Output:** `["a.b.c", "d"]`

## EXPERT (DC18-DC20)

<span style="color:green">**DC18: JSON Transformation**</span>

Transform JSON structure.

**Input:** `{"users": [{"name": "Alice", "age": 30}, {"name": "Bob", "age": 25}]}, transform = extract names`

**Expected Output:** `["Alice", "Bob"]`

<span style="color:green">**DC19: Schema Validation**</span>

Validate against schema.

**Input:** `{"name": "Alice", "age": 30}, schema = {"name": str, "age": int}`

**Expected Output:** `true`

<span style="color:green">**DC20: Custom Dictionary Class**</span>

Implement defaultdict behavior.

**Input:** `dd = DefaultDict(list), dd["a"].append(1), dd["b"].append(2), dd["c"]`

**Expected Output:** `[]` (empty list for missing key)`
