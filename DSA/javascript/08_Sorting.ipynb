## EASY (SO1-SO5)

<span style="color:green">**SO1: Merge Sorted Arrays**</span>
Merge two sorted arrays.

**Output:** `[1, 1, 2, 2, 3, 4]`

<span style="color:green">**SO2: Valid Anagram**</span>
Check if two strings are anagrams using sort.

**Output:** `true`

<span style="color:green">**SO3: Majority Element**</span>
Find element appearing more than n/2 times.

**Output:** `3`

<span style="color:green">**SO4: Sort Colors**</span>
Sort array of 0s, 1s, 2s in place.

**Output:** `[0, 0, 1, 1, 2, 2]`

<span style="color:green">**SO5: Merge Intervals**</span>
Merge overlapping intervals.

**Output:** `[[1, 5]]`
