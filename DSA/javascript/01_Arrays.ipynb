## EASY (A1-A5)

<span style="color:green">**A1: Simple Traversal**</span>
Print all elements in array [1, 2, 3, 4, 5].

**Output:** `1 2 3 4 5`

<span style="color:green">**A2: Sum of Array**</span>
Find sum of [1, 2, 3, 4, 5].

**Output:** `15`

<span style="color:green">**A3: Find Maximum Element**</span>
Find max in [3, 7, 2, 9, 1].

**Output:** `9`

<span style="color:green">**A4: Reverse Array**</span>
Reverse [1, 2, 3, 4, 5].

**Output:** `[5, 4, 3, 2, 1]`

<span style="color:green">**A5: Check if Element Exists**</span>
Check if 7 exists in [1, 3, 5, 7, 9].

**Output:** `true`

## INTERMEDIATE (A6-A12)

<span style="color:green">**A6: Prefix Sum Array**</span>
Create prefix sum array for [1, 2, 3, 4, 5].

**Output:** `[1, 3, 6, 10, 15]`

<span style="color:green">**A7: Sum of Range**</span>
Find sum from index 1 to 3 in [1, 2, 3, 4, 5] using prefix sum.

**Output:** `9` (2+3+4)

<span style="color:green">**A8: Two Sum**</span>
Find two numbers that sum to target. arr=[2, 7, 11, 15], target=9.

**Output:** `[2, 7]` (only one pair)

<span style="color:green">**A9: Three Sum**</span>
Find three numbers that sum to 0 in [-1, 0, 1, 2, -1, -4].

**Output:** `[[-1, -1, 2], [-1, 0, 1]]`

<span style="color:green">**A10: Sliding Window - Max**</span>
Find max in each window of size 3: [1, 3, -1, -3, 5, 3, 6, 7].

**Output:** `[3, 3, 5, 5, 6, 7]`

<span style="color:green">**A11: Container With Most Water**</span>
Find two lines that contain most water. height=[1, 8, 6, 2, 5, 4, 8, 3, 7].

**Output:** `49` (8*7 between indices 1 and 8)

<span style="color:green">**A12: Remove Duplicates**</span>
Remove duplicates from sorted array [1, 1, 2, 2, 3, 3].

**Output:** `[1, 2, 3]`

## ADVANCED (A13-A17)

<span style="color:green">**A13: Trapping Rain Water**</span>
Calculate water trapped. height=[0, 1, 0, 2, 1, 0, 1, 3, 2, 1, 2, 1].

**Output:** `6`

<span style="color:green">**A14: Subarray Sum Equals K**</span>
Count subarrays with sum=5. nums=[1, 2, 2, 3, 1, -3].

**Output:** `3` ([2, 3], [2, 2, 1], [5 via 2+3])

<span style="color:green">**A15: Longest Consecutive Sequence**</span>
Find longest consecutive elements sequence. nums=[100, 4, 200, 1, 3, 2].

**Output:** `4` (sequence: 1, 2, 3, 4)

<span style="color:green">**A16: Rotate Array**</span>
Rotate array k steps. nums=[1, 2, 3, 4, 5], k=2.

**Output:** `[4, 5, 1, 2, 3]`

<span style="color:green">**A17: First Missing Positive**</span>
Find first missing positive. nums=[3, 4, -1, 1].

**Output:** `2`

## EXPERT (A18-A20)

<span style="color:green">**A18: Median of Two Sorted Arrays**</span>
Find median of two sorted arrays. nums1=[1, 3], nums2=[2].

**Output:** `2.0`

<span style="color:green">**A19: Best Time to Buy and Sell Stock**</span>
Find max profit. prices=[7, 1, 5, 3, 6, 4].

**Output:** `5` (buy at 1, sell at 6)

<span style="color:green">**A20: Maximum Product Subarray**</span>
Find max product. nums=[2, 3, -2, 4].

**Output:** `6` (3 * -2 = -6 or 2*3=6)
