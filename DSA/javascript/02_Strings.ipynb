## EASY (S1-S5)

<span style="color:green">**S1: String Reverse**</span>
Reverse the string 'hello'.

**Output:** `'olleh'`

<span style="color:green">**S2: Check Palindrome**</span>
Check if 'racecar' is a palindrome.

**Output:** `true`

<span style="color:green">**S3: Count Character Frequency**</span>
Count frequency of each character in 'hello'.

**Output:** `{h: 1, e: 1, l: 2, o: 1}`

<span style="color:green">**S4: Longest Substring Without Repeating**</span>
Find longest substring without repeating chars in 'abcabcbb'.

**Output:** `3` ('abc')

<span style="color:green">**S5: Check Anagram**</span>
Check if 'listen' and 'silent' are anagrams.

**Output:** `true`
