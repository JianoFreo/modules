## EASY (L1-L5)

<span style="color:green">**L1: Traverse Linked List**</span>
Print all nodes: 1 -> 2 -> 3.

**Output:** `1 2 3`

<span style="color:green">**L2: Reverse Linked List**</span>
Reverse 1 -> 2 -> 3.

**Output:** `3 -> 2 -> 1`

<span style="color:green">**L3: Find Middle of Linked List**</span>
Find middle of 1 -> 2 -> 3 -> 4 -> 5.

**Output:** `3`

<span style="color:green">**L4: Merge Two Sorted Lists**</span>
Merge 1 -> 2 -> 4 and 1 -> 3 -> 4.

**Output:** `1 -> 1 -> 2 -> 3 -> 4 -> 4`

<span style="color:green">**L5: Remove Duplicates**</span>
Remove duplicates from sorted 1 -> 1 -> 2 -> 2.

**Output:** `1 -> 2`

<span style="color:green">**L10: Reverse Nodes in K-Group**</span>

<span style="color:green">**L11: Remove Nth Node**</span>
