## EASY (H1-H5)

<span style="color:green">**H1: Two Sum with Hash Map**</span>
Find two nums summing to 9 in [2, 7, 11, 15].

**Output:** `[2, 7]`

<span style="color:green">**H2: Contains Duplicate**</span>
Check if array has duplicates: [1, 2, 3, 1].

**Output:** `true`

<span style="color:green">**H3: Valid Anagram**</span>
Check if 'anagram' and 'nagaram' are anagrams.

**Output:** `true`

<span style="color:green">**H4: Majority Element**</span>
Find element appearing > n/2 times in [3, 2, 3].

**Output:** `3`

<span style="color:green">**H5: Isomorphic Strings**</span>
Check if 'egg' and 'add' are isomorphic.

**Output:** `true`
