
````python
array = [10, 20, 30, 40, 50]
for i in array:
    print(i)
````

**Q4: Count Occurrences**
Given list = [1, 2, 2, 3, 2, 4], count how many times 2 appears using a loop.

**Expected Output:**
```
3
```

````python
array = [1, 2, 2, 3, 2, 4]
amount = 0
for i in array:
    if i == 2:
        amount += 1
print(amount)
````
