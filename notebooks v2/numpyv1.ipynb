why numpy array are superior

```python
import numpy as np

# Python list - slow
my_list = [1, 2, 3, 4]
my_list = [x * 2 for x in my_list]  # Need loop

# NumPy array - fast
array = np.array([1, 2, 3, 4])
array = array * 2  # Instant, no loop needed
```

```python
import numpy  as np

my_list = [1, 2, 3, 4,]
my_list = my_list * 2

print(my_list)
```

```python
import numpy as np
array = np.array([1, 2, 3, 4])
array = array * 2
print(array)
```

assesment 
Write code to add 5 to every element in a NumPy array [10, 20, 30, 40] and print the result.



```python
import numpy as np 

array = np.array([10, 20, 30, 40])
array = array + 5
print(array)
```

```python
#without numpy
my_list = [10, 20, 30, 40]
my_list = [x + 5 for x in my_list]
print(my_list)  # [15, 25, 35, 45]
```
