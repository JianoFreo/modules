# DOT

```python
import numpy as np
print(np.dot(5, 3))
```

```python
a = np.array([2, 4, 6])
b = np.array([-1, 2, 3])
print(np.dot(a, b))
```

```python
A = np.array([[4, 2, 3, 8],
     [2, 1, 3, 5],
     [1, 1, 2, 6]])

B = np.array([[1, 3],
    [2, 2],
    [8, 2],
    [5, 3]])

print(A)
print(B)
print(np.dot(A, B))
```
