# Why mix integers and strings?

**Real-world use cases:**
- **Data cleaning**: Replace numeric values with categorical labels (e.g., grades: numbers → "PASS"/"FAIL")
- **Reporting**: Convert raw data to human-readable format
- **Missing data**: Replace invalid entries with status strings (e.g., score → "N/A", "INCOMPLETE")

**Key point**: Use `dtype=object` when you need to store mixed types. This trades performance for flexibility since object arrays are slower but can hold anything.

# assesment

Create a student grading system that:

Takes an array of raw test scores: [45, 78, 92, 55, 88, 100, 32, 91]
Converts scores < 60 to "FAIL", scores 60-79 to "PASS", and scores ≥ 80 to "EXCELLENT"
Uses boolean indexing to make these replacements
Returns and prints the final graded array
Constraints:

Use dtype=object so the array can hold strings
Use boolean indexing (like scores[scores < 60]) at least twice
Print the result
This combines: array creation, conditional logic, boolean indexing, and mixed data types.

```python
import numpy as np

# Create integer array for comparisons
scores_int = np.array([45, 78, 92, 55, 88, 100, 32, 91])

# Create result array with dtype=object
result = np.empty(len(scores_int), dtype=object)

# Assign based on conditions (all using the original integer array)
result[scores_int < 60] = "FAIL"
result[(scores_int >= 60) & (scores_int <= 79)] = 'PASS'
result[scores_int >= 80] = 'EXCELLENT'

print(result)
```
