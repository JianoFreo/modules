```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier  
import joblib

music_data = pd.read_csv('music.csv')
x = music_data.drop(columns=['genre']) #drop the genre column
y = music_data['genre']  #set y to be the genre column

model = DecisionTreeClassifier()
model.fit(x, y)# trains the model. the core that reads everything and learns from it
joblib.dump(model, 'music-recommender.joblib')

```
